## Correlation Plots

In version 2 all data is used for plotting the individual scatter plots for each feature

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
file = '../Data/cleaned_data_all_years.csv'
df = pd.read_csv(file)
```

```python
df.head()
```

```python
df.count()
```

```python
clean_data = df.drop(columns=['Country name'])
```

```python
clean_data.count()
```

```python
clean_data.describe().round(2)
```

```python
my_data=clean_data 
```

```python
import seaborn as sns
#create a class instance to pass on the data
g=sns.PairGrid(clean_data,y_vars=['Life Ladder'], x_vars=(['Log GDP per capita',
                                                            'Social support',
                                                            'Healthy life expectancy at birth',
                                                           'Freedom to make life choices',
                                                            'Generosity',
                                                           'Perceptions of corruption',
                                                           'Positive affect',
                                                           'Negative affect']))

# map a function to the grid section
g.map(sns.regplot ,line_kws={'lw':2, 'color':'red'}, scatter_kws={'s':2,'color':'blue'})                                                                
plt.suptitle('Scatter plots for all years', x=0.5 ,y=1, fontsize=20)
plt.tight_layout()
plt.savefig('scatter_plot_overview_linear_all_years')
```

```python
# Individual Scatter Plots
sns.set_style('whitegrid')
g=sns.lmplot(x='Log GDP per capita', y='Life Ladder',data=clean_data, aspect=2,line_kws={"color":"red","alpha":0.7,"lw":2}
            ,scatter_kws={'s':8,'color':'blue'})
g=g.set(xlim=(6,12),ylim=(0,10))
plt.suptitle('Scatter plots:Log GDP percapita', x=0.5 ,y=1, fontsize=20)

plt.savefig('images/Plots/scatter_plot_GDP')
```

```python
# Individual Scatter Plots
sns.set_style('whitegrid')
g=sns.lmplot(x='Social support', y='Life Ladder',data=clean_data, aspect=2,line_kws={"color":"red","alpha":0.7,"lw":2}
            ,scatter_kws={'s':8,'color':'blue'})
g=g.set(xlim=(0.2,1),ylim=(0,9))
plt.suptitle('Scatter plots:Social support', x=0.5 ,y=1, fontsize=20)

plt.savefig('images/Plots/scatter_plot_Social_support')
```

```python
fig, axs = plt.subplots(nrows=4, ncols=2, sharey=False)
plt.rcParams['axes.labelsize'] = 35
plt.suptitle('Scatter plots', x=0.5 ,y=1, fontsize=50)

df.plot(kind='scatter', x='Log GDP per capita', y='Life Ladder', ax=axs[0,0], color='blue', figsize=(30, 30), fontsize=20 )
df.plot(kind='scatter', x='Social support', y='Life Ladder', color='red',ax=axs[0,1], fontsize=20)
df.plot(kind='scatter', x='Healthy life expectancy at birth',y='Life Ladder', color='green',ax=axs[1,0],fontsize=20)
df.plot(kind='scatter', x='Freedom to make life choices', y='Life Ladder', color='purple',ax=axs[1,1], fontsize=20)
df.plot(kind='scatter', x='Generosity', y='Life Ladder', color='deeppink', ax=axs[2,0], fontsize=20)
df.plot(kind='scatter', x='Perceptions of corruption', y='Life Ladder', color='darkred', ax=axs[2,1], fontsize=20)
df.plot(kind='scatter', x='Positive affect', y='Life Ladder', color='navy', ax=axs[3,0], fontsize=20)
df.plot(kind='scatter', x='Negative affect', y='Life Ladder', color='magenta', ax=axs[3,1], fontsize=20)

plt.tight_layout()

plt.savefig('images/Plots/scatter_plot_v2')
plt.show()
```

#### Showing Correlation of the columns using Python's Seaborn

```python
# Heatmap of the Correlation Matrix
import seaborn as sb; sb.set_theme()
import matplotlib as mpl
fig, ax= plt.subplots(1, 1, figsize=(20, 10))
cmap = sb.color_palette('pastel', as_cmap=True)

#cmpap=mpl.cm.cool
#cmap = sb.diverging_palette(, as_cmap=True, n=20)
sb.heatmap(clean_data.corr(), vmin = -1, vmax = 1,annot = True, fmt = ".2f", cmap=cmap)
ax.set_title('Life_Ladder correlation importance')
plt.savefig('heatmap_seaborn_all_years')
```
